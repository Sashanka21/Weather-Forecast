Converting the weather categories into standard categories:

```python
x='Thunderstorms,Moderate Rain Showers,Fog'
```

```python
#making a list of list
list_of_lists=[w.split() for w in x.split(',')]
list_of_lists
```

```python
from itertools import chain #chaaining function together
flat_list=list(chain(*list_of_lists))
flat_list
```

```python
#create a list of this weathers function1
def Create_list(x):
  list_of_lists=[w.split() for w  in x.split(',')]
  flat_list=list(chain(*list_of_lists))
  return flat_list
#convert this weather into a standard list function 2 
def Get_Weather(list1):
  if 'Fog' in list1 and 'Rain' in list1:
    return 'RAIN+FOG'
  elif 'Snow'in list1 and 'Rain'in list1:
    return 'SNOW+FOG'
  elif 'Snow' in list1:
    return 'SNOW'
  elif 'Rain' in list1:
    return 'RAIN'
  elif 'Fog' in list1:
    return 'FOG'
  elif 'Clear' in list1:
    return 'Clear'
  elif 'Cloudy' in list1:
    return 'Cloudy'
  else:
    return 'RAIN'
  

```

```python
Create_list(x)
```

```python
Get_Weather(Create_list(x))
```

```python
data['Std_Weather']=data['Weather'].apply(lambda x:Get_Weather(Create_list(x)))
```

```python
#Importing the Models 
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
```

```python
#Model Initializations
dt_model=DecisionTreeClassifier()
rf_model=RandomForestClassifier()
svc_model=SVC()
knn_model=KNeighborsClassifier()
lr_model=LogisticRegression()
nb_model=GaussianNB()
```

```python
#List Creation
model_list=[dt_model,rf_model,svc_model,knn_model,lr_model,nb_model]
```

```python
#Initializing an empty list
acc_list=[]
#Looping through models
for model in model_list:
    #Training the model
    model.fit(x_train,y_train)
    #Making predictions
    y_pred=model.predict(x_test)
    #Evaluating accuracy
    acc=accuracy_score(y_test,y_pred)
    #Storing accuracy
    acc_list.append(acc)
```

```python
#returning list values
acc_list
```

```python
# creates a Pandas DataFrame
model_df=pd.DataFrame({'Model':model_list,'Accuracy':acc_list})
model_df
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(rf_model,x_std,y,cv=5,scoring='accuracy')
print('Cross-Validation Score=',scores)
```
